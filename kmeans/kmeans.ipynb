```python
import os
os.environ["OMP_NUM_THREADS"] = "1" 

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from tabulate import tabulate
```

```python
# read Data frame pada folder datasets 
df = pd.read_csv("../datasets/Mall_Customers.csv")

# drop kolom customer id karena kurang dibutuhkan
df = df.drop("CustomerID", axis=1)

# show dataset mall_customer
# print("Data dari Dataset")
# print(tabulate(df, headers=["Number", "Genre", "Age","Annual Income", "Spending"])) 

# Seleksi kolom yang akan menentukan cluster
x = df[["Annual Income (k$)", "Spending Score (1-100)"]]

# skalar 
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

print(x_scaled)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(x_scaled)



```

```python
print("Hasil Clustering")
print(tabulate(df, headers=["Number", "Genre", "Age","Annual Income", "Spending", "Cluster"]))
```
